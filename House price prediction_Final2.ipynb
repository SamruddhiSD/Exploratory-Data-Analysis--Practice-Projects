### Handling missing values- categorical features-


```python
missingvalues_categorical = [col for col in nullvalue_columns if data[col].dtypes == 'O']
missingvalues_categorical
```

```python
def replace_missing(data,missingvalues_categorical):
    data1 = data.copy()
    for cols in missingvalues_categorical:
        data1[cols] = data1[cols].fillna(data1[cols].mode()[0])   #replacing null values with respective modes.
    return data1


data = replace_missing(data,missingvalues_categorical)
```

### Performing log normal distribution on the Numerical features (since they are skewed)

```python
data.head()
```

```python
num_features = ['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice']
for col in num_features:
    data[col] = np.log(data[col])
```

```python
data.head()  #notice how the values of the above 5 columns have been transformed now after the logarithmic transformation.Its aimed to transform to normal distribution.
```

```python
cat_features = data.dtypes[data.dtypes == 'object'].index
print(cat_features)
print(len(cat_features))
```

```python
for col in cat_features:
    print('{}:{}'.format(col,data[col].nunique()))
```

```python
def category_onehot_multcols(lst_catcolumns):  #pass the list of categorical columns here as an argument.
    i=0
    for column in lst_catcolumns:
        
        print(column)
        df1=pd.get_dummies(data[column],drop_first=True)  #creating dummies for a column and dropping first column from it.
        
        data.drop([column],axis=1,inplace=True)   #dropping original column containing the categorical feature since we have now converted that to numerical.
        if i==0:
            new_data=df1.copy()
        else:
            
            new_data=pd.concat([new_data,df1],axis=1)  #concatenates previous dataframe of dummy variables with dummy variables of current column
        i=i+1
       
        
    new_data =pd.concat([new_data,data],axis=1)  #concatenates the final dataframe of transformed dummy vars of all the categorical columns and original numeric columns of the original dataframe.
        
    return new_data
```

```python
final_data = category_onehot_multcols(cat_features)
```
